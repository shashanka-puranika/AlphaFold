# Logistic Regression Demo
Train logistic regression and visualize decision boundary.

```python
from pathlib import Path
import sys
import numpy as np
PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
from utils.data_loader import load_csv, train_test_split
from ml_algorithms.logistic_regression import LogisticRegression
from utils.metrics import accuracy, precision, recall
from utils.visualization import plot_decision_boundary
```

```python
df = load_csv(PROJECT_ROOT / 'datasets' / 'synthetic_classification.csv')
x = df[['feature1', 'feature2']].to_numpy(dtype=float)
y = df['label'].to_numpy(dtype=int)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_seed=42)
model = LogisticRegression(learning_rate=0.1, n_epochs=1200)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print('Accuracy:', accuracy(y_test, y_pred))
print('Precision:', precision(y_test, y_pred))
print('Recall:', recall(y_test, y_pred))
```

```python
plot_decision_boundary(model, x_test, y_test, title='Logistic Regression Decision Boundary')
```
